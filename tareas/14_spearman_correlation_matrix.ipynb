
# Actividad 14: Matriz de correlación de Spearman

Se calcula y grafica la matriz de correlación de Spearman para salinidad, TA y DIC.


```python
import pandas as pd

# Leer dataset
CO2Data = pd.read_csv("../data/Terminos_lagoon_TA_DIC_2023_RawData.csv")

# Seleccionar columnas de interés
ta_sal_dic_df = CO2Data[['sal_psu', 'ta_micromol_kg', 'dic_micromol_kg']]

# Revisar las primeras filas
print(ta_sal_dic_df.head())

# Calcular la matriz de Spearman
corr_matrix = ta_sal_dic_df.corr(method='spearman')
print(corr_matrix)

# Graficar heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.set(style="white", font_scale=1.2)

sns.heatmap(
    corr_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap="cividis", 
    square=True, 
    linewidths=0.5, 
    linecolor="gray",
    cbar_kws={"shrink": 0.8}
)

plt.title("Spearman Correlation Matrix", fontsize=14)
plt.tight_layout()
plt.show()
```

## Ejercicio adicional de práctica

Además de la matriz de correlación original, quise hacer otra incluyendo la variable de temperatura (temp_c).  
La idea es practicar más el uso de la matriz de Spearman y ver cómo se relaciona la temperatura con salinidad, TA y DIC.


```python
# Selección de variables incluyendo temperatura
ta_sal_dic_temp_df = CO2Data[['sal_psu', 'ta_micromol_kg', 'dic_micromol_kg', 'temp_c']]

# Revisar primeras filas
print(ta_sal_dic_temp_df.head())

# Calcular matriz de Spearman
corr_matrix_temp = ta_sal_dic_temp_df.corr(method='spearman')
print(corr_matrix_temp)

# Volvemos a importar por si esta celda corre aislada
import matplotlib.pyplot as plt
import seaborn as sns

# Graficar heatmap
plt.figure(figsize=(8, 6))
sns.set(style="white", font_scale=1.2)

sns.heatmap(
    corr_matrix_temp, 
    annot=True, 
    fmt=".2f", 
    cmap="cividis", 
    square=True, 
    linewidths=0.5, 
    linecolor="gray",
    cbar_kws={"shrink": 0.8}
)

plt.title("Spearman Correlation Matrix (incluyendo temperatura)", fontsize=14)
plt.tight_layout()
plt.show()
```
