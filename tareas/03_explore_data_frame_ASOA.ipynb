# Exploración de pandas y data frames


### Exportar el data frame

```python
import pandas as pd
RUTA = "../data/Terminos_lagoon_TA_DIC_2023_RawData.csv"
lagoon_df = pd.read_csv(RUTA)

lagoon_df.head()
```

### Exploración del data frame


```python
# Imprimir nombres de las variables
print(lagoon_df.columns)
```

#### Utilización de la variable sal para realizar la exploración

```python
# Guardar la variable de salinidad
sal = lagoon_df['sal_psu']
```

```python
# Graficar la variable salinidad
sal.plot(title="Salinidad (PSU)")
```

```python
# Subconjunto: sólo datos de temporada lluviosa
sal_rainy = lagoon_df[lagoon_df['season'] == 'Rainy']
sal_rainy
```

```python
# Subconjunto: excluir estuario Candelaria
sal_no_Candelaria = lagoon_df[lagoon_df['estuary'] != 'Candelaria']
sal_no_Candelaria
```

```python
# Subconjunto: muestras de superficie (>0.5 m de profundidad)
sal_surface = lagoon_df[lagoon_df['depth_m'] > 0.5]
sal_surface
```
