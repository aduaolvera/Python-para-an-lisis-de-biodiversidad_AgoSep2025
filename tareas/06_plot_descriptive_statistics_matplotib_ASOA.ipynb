# Explorar datos ambientales usando Pandas

```python
import pandas as pd

# Importar dataset ambiental
CO2Data = pd.read_csv("../data/Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(CO2Data.info())
```

## Histograma simple de salinidad

```python
import matplotlib.pyplot as plt

# Seleccionar datos de salinidad
sal_data = CO2Data["sal_psu"]

plt.hist(sal_data, bins=10)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Frequency")
plt.title("Histogram of Salinity")
plt.show()
```

## Histograma normalizado (densidad)

```python
# Histograma normalizado (density=True)
data = CO2Data["sal_psu"]

plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probability")
plt.title("Normalized Histogram of Salinity")
plt.show()
```

## Boxplot de oxígeno disuelto

```python
do_data = CO2Data["do_mg_l"]

plt.boxplot(do_data)
plt.xlabel("DO")
plt.ylabel("Concentration (mg/L)")
plt.title("Boxplot of DO Concentration")
plt.show()
```

## Boxplot de salinidad con notch

```python
data = CO2Data["sal_psu"]

fig2, ax2 = plt.subplots()
ax2.set_title("Notched box for Salinity")
ax2.boxplot(data, notch=True)

plt.ylabel("Salinity (PSU)", fontsize=12)
plt.show()
```

## Boxplot de TA con símbolos de outliers cambiados

```python
data = CO2Data["ta_micromol_kg"]

green_diamond = dict(markerfacecolor="g", marker="D")
fig3, ax3 = plt.subplots()

ax3.set_title("Boxplot of TA (changed outlier symbols)")
ax3.boxplot(data, flierprops=green_diamond)

plt.ylabel("TA ($\\mu mol \; kg^{-1}$)", fontsize=12)
plt.show()
```

# Ejercicio: histogramas de DIC y TA

```python
# Histograma de DIC
dic_data = CO2Data["dic_micromol_kg"]

plt.hist(dic_data, bins=10)
plt.xlabel("DIC ($\\mu mol \; kg^{-1}$)")
plt.ylabel("Frequency")
plt.title("Histogram of DIC")
plt.show()

# Histograma de TA
ta_data = CO2Data["ta_micromol_kg"]

plt.hist(ta_data, bins=10)
plt.xlabel("TA ($\\mu mol \; kg^{-1}$)")
plt.ylabel("Frequency")
plt.title("Histogram of TA")
plt.show()
```
